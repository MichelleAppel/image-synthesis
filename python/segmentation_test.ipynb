```python
import torch
from torchvision.utils import make_grid
from synthesis_dataset import SynthesisDataset
from torch.utils.data import DataLoader
```

```python
torch.manual_seed(42)

synthesis_dataset = SynthesisDataset("..\..\simulation-synthesis\output\MLDataset_128", extension='.png')
loader = DataLoader(synthesis_dataset, shuffle=True, drop_last=True, batch_size = 4)
it = iter(loader)
```

```python
batch = it.next()
idimg = batch['indexid']
```

```python
synthesis_dataset.toPilImage(make_grid(batch['img']))
```

```python
synthesis_dataset.toPilImage(make_grid(batch['class']).float())
```

```python
synthesis_dataset.toPilImage(make_grid(synthesis_dataset.class_to_color(batch['class'])))
```

```python
classimg = batch['class']
synthesis_dataset.toPilImage(make_grid(classimg == 0)*1.)
```

```python
for c in idimg[classimg == 0].unique():
    print(synthesis_dataset.meta[c.item()]['objectname'], synthesis_dataset.meta[c.item()]['parentsname'])
```
